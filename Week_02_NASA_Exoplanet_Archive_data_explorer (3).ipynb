### 3, Research on the state-of-the-art performance of each detection technique and overplot the sensitivity line. An example of RV is given below.

#### $K = \frac{m_p}{m_\ast}\cdot\sqrt{\frac{Gm_\ast}{a}}\cdot\sin i$

#### Assuming an edge-on orbit, for which $i$ is 90 degree, rearrange the equation to be:

#### $m_p = K \cdot m_\ast \cdot \sqrt{\frac{a}{Gm_\ast}}$

#### For a K value of 0.5 m/s, which is the state-of-the-art performance, the mass sensitivity is as follows for a range of semi major axis $a$, assuming 0.5 solar mass:

```python
import astropy.constants as ac
import astropy.units as u
```

```python
sma_arr = 10.0**(np.arange(5)-2) * u.AU

m_p_arr = 0.5 * u.meter / u.second * 0.5 * u.solMass * np.sqrt(sma_arr / ac.G / (0.5 * u.solMass))
```

```python
plt.plot(data['pl_orbsmax'], data['pl_bmasse'], 'k.', label='Exoplanet Data')
plt.plot(ss_semimaj, ss_masse, 'r.', label = 'Solar System Data')
plt.plot(sma_arr.value, m_p_arr.decompose().to(u.earthMass), color='blue', label='Sensitivity Line')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Semi-Major Axis [AU]')
plt.ylabel('Mass [R$_{Earth}$]')
plt.legend()
plt.savefig('fig7.png', dpi=300)
```

## II, Calculating Detection Signal of Different Methods

### For a Earth-like planet around a Sun-like star

(RV, transiting depth, duration, planet-star flux ratio in reflection and thermal emission, astrometric signal)

```python
#radial velocity function
def rad_vel(mp, ms, a):
    """
    Required arguments:
       mp       mass of planet in kg
       ms       mass of star in kg 
       a        semi-major axis in meters
    """
    G = 6.67408e-11 #m3 kg-1 s-2
    r = (mp / ms) * (G * ms / a)**(1/2)
    return r

print(rad_vel(5.972e24, 1.989e30, 149597887500), 'm/s')
```

```python
#transitioning depth function
def trans_depth(rp, rs):
    """
    Required arguments:
       rp       radius of planet
       rs       radius of star
    """
    f = (rp / rs)**2
    return f

print(trans_depth(63781.1, 696347.055), 'km')
```

```python
#transit duration function
def duration(rs, vk):
    """
    Required arguments:
       rs       radius of star
       vk       velocity in m/s
    """
    t = (2 * rs / vk)
    return t

print(duration(63781.1, 300), 's')
```

```python
#planet-star flux ratio in reflection and thermal emission
#def ps_flux()
    
    
def therm(v, T):
    h = 6.62607004e-34 #m2 kg / s
    c = 3.0e8 #m/s
    k = 1.38064852e-23 #m2 kg s-2 K-1
    z = (h * v) / (k * T)
    B = ((2 * h * v**3) / c**2) * 1 / ((np.exp(z)) - 1)
    return B

print(therm(.00001157407, 288), therm(4.82160077e-7, 3500))
```

```python
#astrometric signal
def astro_sig(mp, ms, a, d):
    """
    Required arguments:
       mp      mass of planet
       ms      mass of star
       a       osemi-major axis in meters
       d       distance (choose an average dist)
    """
    θ = (mp / ms) * (a / d)
    return θ

print(astro_sig(5.972e24, 1.989e30, 4.84814e-6, 400), 'rad')
```
